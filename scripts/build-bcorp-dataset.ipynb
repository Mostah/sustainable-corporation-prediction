```python
import pandas as pd

df = pd.read_excel('../data/raw/initial_data.xlsx')
df.head()
```

```python
print(df['size'].unique())

sizes_to_keep = [' 250-999', ' 50-249', ' 250+', ' 1000+', ' 10,000+']
big_companies = df[df['size'].isin(sizes_to_keep)]
big_companies.head()
```

```python
big_companies.sort_values('date_certified', ascending=False, inplace=True)
big_companies.drop_duplicates('company_id', inplace=True)
big_companies.head()
```

```python
big_companies.to_pickle('../data/processed/b_corporation_scores.pkl')
```
